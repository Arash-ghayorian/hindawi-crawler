```python
import requests
from bs4 import BeautifulSoup
```

```python
list_links = []
```

```python
# 170 page
for x in range(1, 170):
    page = requests.get(f'https://www.hindawi.org/books/{x}/')
    soup = BeautifulSoup(page.text, 'html.parser')
    books = soup.find_all(class_='bookCover')
    for i in books:
        list_links.append('https://www.hindawi.org' + str(i.find_all('a')[0]).split('<a href="')[1].split('/">')[0])
```

```python
fil = open(r'C:\Users\lapsus\Desktop\hindawi\links.txt', 'r')
lines = fil.readlines()
for i in lines:
    list_links.append(i.split('\n')[0])
```

```python
list_links
```

```python
import pandas as pd
df = {
    'book_name': [],
    'author': [],
    'tags': [],
    'safhe': [],
    'kholase_ketab': [],
    'book_link_pdf': [],
    'firs_publish': [],
    'translator': []
}

for i in list_links:
    page = requests.get(i)
    soup = BeautifulSoup(page.text, 'html.parser')
    df['book_name'].append(soup.find_all(class_='details')[0].find_all('h2')[0].text)
    try:
        df['author'].append(soup.find_all(class_='author')[0].find_all('a')[0].text)
    except:
        df['author'].append(None)
    try:
        df['translator'].append(soup.find_all(class_='meta')[0].find_all('a')[0].text)
    except:
        df['translator'].append(None)
    df['tags'].append(soup.find_all(class_='tags')[0].find_all('a')[0].text.split('                                    ')[1].split('\r\n')[0])
    df['safhe'].append(soup.find_all(class_='tags')[0].find_all('span')[0].text.split('كلمة  ')[0].split(' ')[1])
    df['kholase_ketab'].append(soup.find_all(class_='content')[0].text.replace('\n', ' ').replace('\r                              ', ' '))
    df['book_link_pdf'].append(str(soup.find_all(class_='downloadBook')[0]).split('PDF" href="')[1].split('" id="pdf"')[0])
    try:
        df['firs_publish'].append(soup.find_all(class_='releaseDates')[0].find_all('spn')[0].text.replace(' ', ''))
    except:
        df['firs_publish'].append(soup.find_all(class_='releaseDates')[0].find_all('li')[0].text.replace('  ', '').split(' ')[7].split('.')[0])

```
